```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt

personID = [100,101,102,103,104,105,106,107,108,109,110,111,112,113,114,115,116,117,118,119,120]
TV = [2,.5,6,3,4,2,0,.5,3,0,2.5,3.5,2,6,4,1.5,2,3,7,4,0]
VideoGames = [1.5,3,0,0,.5,0,5,3,2,.5,.5,0,3,4,1.5,1,2,2.5,3.5,1.5,0]
Reading = [0,0,.5,0,1.5,.5,.5,0,0,3,0,.5,1,.5,1.5,0,4,0,2,1,.5]
Sleeping = [9,8.5,10,7.5,9,8.5,7,9.5,8,8,8.5,11,10,6.5,8,8.5,9,9,8,9.5,8]
Date = ["06/02/2017","06/05/2017","07/11/2017","06/16/2017","06/25/2017","06/01/2017","06/19/2017","06/30/2017","07/03/2017","07/06/2017","06/12/2017","06/08/2017","06/23/2017","06/11/2017","06/25/2017","07/02/2017","06/24/2017","06/03/2017","06/09/2017","06/18/2017","07/03/2017"]

```

```python
#Creating a new DF
timeDF = pd.DataFrame(list(zip(personID,TV,VideoGames, Reading, Sleeping, Date)),columns=["personID","TV","VideoGames","Reading", "Sleeping", "Date"])
display(timeDF)
```

```python
#Displaying data types
timeDF.info()
```

```python
#Converting to datetime format
timeDF["Date"] = pd.to_datetime(timeDF["Date"])
#Datetime is not int/float/etc.
```

```python
timeDF.info()
```

```python
#Assigning specific date
date = dt.datetime(2017,7,1)
print(date)
```

```python
#Finding current time now
currentTime = dt.datetime.now()
print(currentTime)
```

```python
#Displaying only dates that occur before July 1st
timeDFJune = timeDF[timeDF["Date"] < date]
display (timeDFJune)
```

```python
#Sorting from earliest to latest date
timeDFJune = timeDFJune.sort_values(by="Date")
display(timeDFJune)
```

```python
#Creating a new column that adds up all time spent on all activities
timeDFJune["All_Time_Spent"] = timeDFJune["TV"] + timeDFJune["VideoGames"] + timeDFJune["Reading"] + timeDFJune["Sleeping"]
display(timeDFJune)
```

```python
#Displaying all time unaccounted for
timeDFJune["Unrecorded_Time_Spent"] = 24 - timeDFJune["All_Time_Spent"]
display(timeDFJune)
```

```python
timeDFJune["Waking_Time_Spent"] = (timeDFJune["TV"] / (24 - timeDFJune["Sleeping"])) * 100
display(timeDFJune)
```
